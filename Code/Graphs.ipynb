```python
figure(num=None, figsize=(12, 6), dpi=80, facecolor='w', edgecolor='k')

# set width of bar
barWidth = 0.25
 
# set height of bar
onehot = [100, 93.01, 57.35, 22.17, 5.78, 1.44]
similarity = [100, 100, 94.94, 48.43, 15.42, 9.64]
word_embedding = [100, 95.43, 89.64, 78.07, 69.64, 50.6]
bert = [100, 32.29, 24.3, 17.83, 13.49, 12.05]

onehot_std = [0, 2.47, 5.36, 7.6, 1, 1.57]
similarity_std = [0, 0, 1.32, 4.21, 3.94, 2.55]
word_embedding_std = [0, 1.32, 2.19, 2.88, 1.79, 2.09]
bert_std = [0, 1.98, 2.19, 1.32, 1.57, 0]

# Set position of bar on X axis
r1 = np.arange(len(onehot))
r2 = [x + 0.7*barWidth for x in r1]
r3 = [x + 0.7*barWidth for x in r2]
r4 = [x + 0.7*barWidth for x in r3]
 
# Make the plot
plt.bar(r1, onehot, width=0.7*barWidth, yerr=onehot_std, capsize=5, edgecolor='white', label='One-Hot', color = "cyan")
plt.bar(r2, similarity, width=0.7*barWidth, yerr=similarity_std, capsize=5, edgecolor='white', label='Similarity', color = "orange")
plt.bar(r3, word_embedding, width=0.7*barWidth, yerr=word_embedding_std, capsize=5, edgecolor='white', label='Word Values', color = "green")
plt.bar(r4, bert, width=0.7*barWidth, yerr=bert_std, capsize=5, edgecolor='white', label='BERT', color = "red")
 
# Add xticks on the middle of the group bars
plt.xlabel('Noise Level', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(onehot))], ['0%', '10%', '20%', '30%', '40%', '50%'])
plt.ylabel('Mean accuracy (%)', fontweight='bold')
 
# Create legend & Show graphic
plt.title("Accuracy comparison, non-complement coding retrieval")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
