```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
 
# set height of bar
onehot = [100, 93.01, 57.35, 22.17, 5.78, 1.44]
similarity = [100, 100, 94.94, 48.43, 15.42, 9.64]
word_embedding = [100, 95.43, 89.64, 78.07, 69.64, 50.6]
 
# Set position of bar on X axis
r1 = np.arange(len(onehot))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, onehot, width=barWidth, edgecolor='white', label='One-Hot')
plt.bar(r2, similarity, width=barWidth, edgecolor='white', label='Similarity')
plt.bar(r3, word_embedding, width=barWidth, edgecolor='white', label='Word Embedding')
 
# Add xticks on the middle of the group bars
plt.xlabel('Noise Level', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(onehot))], ['0%', '10%', '20%', '30%', '40%', '50%'])
plt.ylabel('Mean accuracy (%)', fontweight='bold')
 
# Create legend & Show graphic
plt.title("Accuracy comparison, non-complement coding retrieval")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
